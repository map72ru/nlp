```python
# Загружаем данные
data = open('hw2/MyDrive/data/corpus').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

```python
from sklearn import model_selection, preprocessing, linear_model
from sklearn.metrics import accuracy_score
```

```python
def calculate(vectorizer, trainDF):
  train_x, valid_x, train_y, valid_y = model_selection.train_test_split(trainDF['text'], trainDF['label'])

# labelEncode целевую переменную
  encoder = preprocessing.LabelEncoder()
  train_y = encoder.fit_transform(train_y)
  valid_y = encoder.fit_transform(valid_y)


  vectorizer.fit(trainDF['text'])
  xtrain_count =  vectorizer.transform(train_x)
  xvalid_count =  vectorizer.transform(valid_x)

  classifier = linear_model.LogisticRegression(solver='lbfgs', max_iter=1000)
  classifier.fit(xtrain_count, train_y)
  predictions = classifier.predict(xvalid_count)
  return accuracy_score(valid_y, predictions)
```

```python
count_vect = CountVectorizer(analyzer='word', token_pattern=r'\w{1,}')
result = calculate(count_vect, trainDF)
print(result)
```

```python
def replacer(value: str, vocabulary: dict) -> str:
  word = value
  for token in value.split():
    if token in vocabulary.keys():
       word = word.replace(token, vocabulary[token])
  return word
```

```python
import re
data = trainDF.copy()
data['text'] = data.apply(lambda x: x['text'].lower(), axis=1)
data['text'] = data.apply(lambda x: replacer(x['text'], apostrophe_dict), axis=1)
data['text'] = data.apply(lambda x: replacer(x['text'], short_word_dict), axis=1)
data['text'] = data.apply(lambda x: replacer(x['text'], emoticon_dict), axis=1)
data['text'] = data.apply(lambda x: re.sub(r'[^\w\s]', ' ', x['text']), axis=1)
data['text'] = data.apply(lambda x: re.sub(r'[^a-zA-Z0-9]', ' ', x['text']), axis=1)
data['text'] = data.apply(lambda x: re.sub(r'[^a-zA-Z]', ' ', x['text']), axis=1)
data['text'] = data.apply(lambda x: ' '.join([w for w in x['text'].split() if len(w)>1]), axis=1)
```

```python
nltk.download('wordnet')
data1 = data.copy()
ps = nltk.stem.wordnet.WordNetLemmatizer()
data1['text'] = data1.apply(lambda x: [ps.lemmatize(w) for w in x['text'].split()], axis=1)
```
